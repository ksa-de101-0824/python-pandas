```python
import pandas as pd
```

### Read Multiple Files - Pizza Sales and Order

```python
orders = pd.read_csv('./Pizza Sales/orders.csv', header=0)
order_details = pd.read_csv('./Pizza Sales/order_details.csv', header=0)
```

```python
pizza = pd.read_csv('./Pizza Sales/pizzas.csv', header = 0)
pizza_types = pd.read_csv('./Pizza Sales/pizza_types.csv', header = 0, encoding='1252')
```

### Join all 4 data

```python
# df.merge()
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.


orders\
    .merge(order_details, how='left', left_on = 'order_id', right_on = 'order_id')\
    .merge(pizza, how='left', on='pizza_id')\
    .merge(pizza_types, how='left', on='pizza_type_id')\
    .sample(10)
```

Then we can perform visualization, filtering, exporting like previous steps...
