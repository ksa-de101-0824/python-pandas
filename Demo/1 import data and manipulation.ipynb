### Filtering Data

#### Method 1: Using .loc[]

*Format example*:

```df.loc[ df['column_name'] >= .... ]``` Filter on column_name with value more than equal to ....

```df.loc[ (df['column_name1'] >= ....) & (df['column_name2'] != ....)]``` Filter on column_name1 with value more than equal to .... AND filter on column_name2 with value not equal to ....

*Boolean operators*:
1. | = represents OR
2. & = represents AND
3. ~ = represents NOT
 https://pandas.pydata.org/docs/user_guide/indexing.html#boolean-indexing

````python
# Filter data where is_arrested == True
df_crime_only_arrested = df_crime.loc[df_crime['primary_type'] == 'BATTERY']
df_crime_only_arrested.head(3)
````

````python
# Filter data where a column of string values containing a word like "domestic"
df_crime.loc[df_crime['description'].str.contains('domestic', case=False)]
````

````python
# Find out how much % of crime reported was arrested?
print(df_crime_only_arrested.shape)
print(df_crime.shape)

# Calculate the percentage of arrest out of total
total_arrested = df_crime_only_arrested.shape[0]
total_crime_recorded = df_crime.shape[0]

print(f"Total arrested was {total_arrested} out of {total_crime_recorded}")
print(f"Only {total_arrested * 100 / total_crime_recorded}% of reported crime was arrested")
````

````python
# Find out how much % of crime reported was arrested, excluding "Theft"
# In this case, filter only Arrest = True & Primary Type = "Theft"
df_crime_only_non_theft_arrested = df_crime.loc[(df_crime.arrested == True) & (df_crime['primary_type'] != ' BATTERY')]
df_crime_only_non_theft_arrested.sample(5)
````

````python
df_crime_only_non_theft_arrested.shape[0] * 100 / df_crime.shape[0]
````

#### Method 2: Using .agg()

<img src = "images/groupby1.png">

https://keytodatascience.com/groupby-pandas-python/

````python
# Get total crime by primary_crime_type
df_crime_to_solve = df_crime\
                        .groupby( ['primary_crime_type']  ) \
                        .agg(
                            total_crime = ('unique_id', 'count'),
                            total_arrests = ('is_arrested', 'sum')
                            )\
                        .reset_index()
    
df_crime_to_solve
````

````python
import seaborn as sns
# Besides matplotlib, we can plot using Seaborn too
# https://seaborn.pydata.org/generated/seaborn.barplot.html
sns.barplot( 
            data = df_crime_to_solve,
            x = 'primary_crime_type',
            y = 'total_crime',
            color = 'purple'
            )

plt.xticks(rotation=90);
````

````python
# Sort the plot by totals of crime, make it nicer
plt.figure(figsize = (12,3)) # Prepare the canvas, specify the size (optional)
plt.title('Total Crime by Type') # Write on the canvac a title

# Plot the graph on canvas
sns.barplot( 
        data = df_crime_to_solve.sort_values('total_crime', ascending=False),
        x = 'primary_crime_type',
        y = 'total_crime',
        )       
# Adjust rotation of x-label
plt.xticks(rotation=60);
````
